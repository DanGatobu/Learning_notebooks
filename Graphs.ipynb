```python
def getcollum(name,dataframe):
    name=name
    if  isinstance(name,list) :
        da=dataframe[name]
    else:
        dtains=str(type(name))
        print(f"takes in list you've entered {dtains} type  ")
    return da
```

```python
collums=['alcohol']
qual=['quality']
s=getcollum(collums,df3)
why=getcollum(qual,df3)
why
```

```python
def spli(xdata,ydata):
    x_train,x_test,y_train,y_test=train_test_split(xdata,ydata,test_size=0.33,random_state=42)
    return x_train,x_test,y_train,y_test

def norm(data):
    scale=StandardScaler()
    scale.fit(data)
    da=scale.transform(data)
    return da
```

```python
x_train,x_test,y_train,y_test=spli(s,why)
normxtrain=norm(x_train)
normytrain=norm(y_train)
normxtest=norm(x_test)
normytest=norm(y_test)


```

```python
def makemodel(xtrain,ytrain,xtest,ytest):
    model = DecisionTreeRegressor(random_state=42)
    model.fit(xtrain,ytrain )
    y_pred = model.predict(xtest)
    mse = mean_squared_error(ytest, y_pred)
    rmse = np.sqrt(mse)
    print(f'Mean Squared Error: {mse}')
    print(f'Root Mean Squared Error: {rmse}')
    return model,mse,rmse

    
```

```python
model=makemodel(normxtrain,normytrain,normxtest,normytest)
```
