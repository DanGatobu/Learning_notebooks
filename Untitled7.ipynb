```python
%matplotlib inline 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Read the data from CSV files
songs_2010 = pd.read_csv('2010.csv')
songs_2000 = pd.read_csv('2000.csv')
songs_1990 = pd.read_csv('1990.csv')
songs_1980 = pd.read_csv('1980.csv')
songs_1970 = pd.read_csv('1970.csv')
songs_1960 = pd.read_csv('1960.csv')

# Concatenate all the data into a single DataFrame
allsongs = pd.concat([songs_2010, songs_2000, songs_1990, songs_1980, songs_1970, songs_1960], axis=0, ignore_index=True)

# Extract the required features
allsongs_bpm = allsongs['bpm']
allsongs_nrgy = allsongs['nrgy']
allsongs_dnce = allsongs['dnce']

# Combine the features into a DataFrame
data = pd.DataFrame({'bpm': allsongs_bpm, 'nrgy': allsongs_nrgy, 'dnce': allsongs_dnce})

# # Choose the number of clusters
# num_clusters = 4

# # Create an instance of KMeans and fit the data
# kmeans = KMeans(n_clusters=num_clusters)
# kmeans.fit(data)

# # Get the cluster labels and add them to the DataFrame
# data['cluster'] = kmeans.labels_

# # Visualize the clusters
# plt.scatter(data['bpm'], data['nrgy'], c=data['cluster'], s=data['dnce'])
# plt.xlabel('BPM')
# plt.ylabel('Energy')
# plt.show()
import numpy as np

def kmeans_manual(data, num_clusters, max_iterations=100):
    # Randomly initialize cluster centers
    centroids = data[np.random.choice(range(data.shape[0]), size=num_clusters, replace=False)]

    for _ in range(max_iterations):
        # Assignment step: Assign each data point to the nearest cluster center
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
        cluster_labels = np.argmin(distances, axis=1)

        # Update step: Compute new cluster centers as the mean of the data points in each cluster
        new_centroids = np.array([data[cluster_labels == k].mean(axis=0) for k in range(num_clusters)])

        # Check for convergence
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return cluster_labels, centroids

# Usage example
data = np.column_stack((allsongs_bpm, allsongs_nrgy, allsongs_dnce))
num_clusters = 4

cluster_labels, centroids = kmeans_manual(data, num_clusters)

# Visualize the clusters
plt.scatter(data[:, 0], data[:, 1], c=cluster_labels)
plt.xlabel('BPM')
plt.ylabel('Energy')
plt.show()
```
